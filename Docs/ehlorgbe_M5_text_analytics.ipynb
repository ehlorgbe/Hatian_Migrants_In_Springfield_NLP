# Natural Language Processing
Textual data is unstructured data. Structured data have a predetermined format. They are usually organized into records with simple data values (categorical, ordinal, and continuous variables) and stored in databases. In contrast, unstructured data do not have a predetermined format and are stored in the form of textual documents. There are some methods for processing textual data.  The Natural Language Toolkit (NLTK) can be used for various data processing techniques such as tokenizing, tagging parts of speech, extracting phrases, stemming and sentiment analysis.

## Tokenizing and tagging

## Sentiment

```python
import nltk
nltk.download('vader_lexicon')

from nltk.sentiment import SentimentIntensityAnalyzer

# right wing article
analyzer = SentimentIntensityAnalyzer()
scores = analyzer.polarity_scores(article1)
neg = scores['neg']
pos = scores['pos']
neutral = scores['neu']
if neg > pos :
  print("Negative sentiment")
elif pos > neg :
  print("Positive sentiment")
else:
  print("Neutral sentiment")
print(scores)

"""
You’ll get back a dictionary of different scores. The negative, neutral, and
positive scores are related: They all add up to 1 and can’t be negative. The
compound score is calculated differently. It’s not just an average, and it
can range from -1 to 1.
"""
```
